Using the CMIP6 database hosted on Amazon Web Services, choose any two models you like: use both their historical simulations and future projections from one of the four major SSPs (ssp126, ssp245, ssp370, or ssp585). Provide a brief description of the models and scenarios you chose to include. 

```python
# Import libraries
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import intake
```

I am choosing the CanESM5 and CESM models. CESM5 is the Canadian Earrth System Model version 5 that is made up of 3 dimensional atmospheric and oceanic genereal circulation models (Swart et al). CESM is the community earth system model. I will be using the high emission scenario SSP370.

```python
# Open the CMIP6 data catalog, store as a variable
catalog = intake.open_esm_datastore('https://cmip6-pds.s3.amazonaws.com/pangeo-cmip6.json')


# member_id: which ensemble member do you want?
member_id = 'r10i1p1f1'

# table_id: which part of the Earth system and time resolution do you want? Here we want monthly atmosphere data
table_id = 'Amon'

# variable_id: which climate variables do you want? Here we want surface air temperature
variable_id = ['tas']
```

```python
# Historical
can_hist = catalog.search(
    activity_id='CMIP',
    source_id='CanESM5',
    experiment_id='historical',
    member_id=member_id,
    table_id=table_id,
    variable_id=variable_id
)


# SSP370
can_fut = catalog.search(
    activity_id='ScenarioMIP',
    source_id='CanESM5',
    experiment_id='ssp370',
    member_id=member_id,
    table_id=table_id,
    variable_id=variable_id
)


# Historical CESM
cesm_hist = catalog.search(
    activity_id='CMIP',
    source_id='CESM2',
    experiment_id='historical',
    member_id=member_id,
    table_id=table_id,
    variable_id=variable_id)

# SSP370 Can
cesm_fut = catalog.search(
    activity_id='ScenarioMIP',
    source_id='CESM2',
    experiment_id='ssp370',
    member_id=member_id,
    table_id=table_id,
    variable_id=variable_id
)

```

## Using Atmospheric surface air temperature

```python
# Open xarray
hist_cesm = xr.open_zarr(cesm_hist.df.iloc[0]['zstore'], storage_options={'anon': True}, consolidated=True)
future_cesm = xr.open_zarr(cesm_fut.df.iloc[0]['zstore'], storage_options={'anon': True}, consolidated=True)
hist_can = xr.open_zarr(can_hist.df.iloc[0]['zstore'], storage_options={'anon': True}, consolidated=True)
future_can = xr.open_zarr(can_fut.df.iloc[0]['zstore'], storage_options={'anon': True}, consolidated=True)

# 
hist_cesm['time'] = hist_cesm.time.astype('datetime64[ns]')
future_cesm['time'] = future_cesm.time.astype('datetime64[ns]')
hist_can['time'] = hist_can.time.astype('datetime64[ns]')
future_can['time'] = future_can.time.astype('datetime64[ns]')


# Concatenate
cesm_data = xr.concat([hist_cesm, future_cesm], dim="time")
can_data = xr.concat([hist_can, future_can], dim="time")

# Sort by time for future filtering 
can_data = can_data.sortby('time')
cesm_data = cesm_data.sortby('time')
```

## Select Region

I will be using Alaska as my region. Located in the artic circle, it is a highly sensitive region to climatic changes due to drastically changing temperatures. Habitats and species are specialized for this region, making the environment such a sensitive region that is susceptible to melting ice caps and species range shifts.

```python
alaska_bbox = [-179.13657211802118, 51.214183,179.77488070600702,71.365162]
```

```python
# Define time periods (historical and projected)
hist = [1980, 2020]
proj = [2050, 2100]

# Filter data of each model to historical and future time periods 
# CanESM 
can_tas_hist = can_data.sel(time=slice(str(hist[0])+"-01-01", str(hist[1])+"-12-31"))
can_tas_proj = can_data.sel(time=slice(str(proj[0])+"-01-01", str(proj[1])+"-12-31"))

# MIROC miroc_tas
cesm_tas_hist = cesm_data.sel(time=slice(str(hist[0])+"-01-01", str(hist[1])+"-12-31"))
cesm_tas_proj = cesm_data.sel(time=slice(str(proj[0])+"-01-01", str(proj[1])+"-12-31"))


# Calc averages over time periods and convert to celcius 
can_tas_histmn = can_tas_hist.mean(dim="time") - 273.15
can_tas_projmn = can_tas_proj.mean(dim="time") - 273.15

cesm_tas_histmn = cesm_tas_hist.mean(dim="time") - 273.15
cesm_tas_projmn = cesm_tas_proj.mean(dim="time") - 273.15
```

I chose 1980-2020 for the historical time periods to get a good look at the data from the past 40 years. Projecting forward, I wanted to see what the second half of the century is projected to look like at this level of emissions at SSP 370.

### CanESM5 historical map - SSP370

```python
# Define object containing PlateCarree projection
map = ccrs.PlateCarree()

# Make figure and axes
fig, ax = plt.subplots(figsize=(20, 10), subplot_kw={"projection": map})

# Plot 
plot = ax.pcolormesh(can_tas_histmn.lon, can_tas_histmn.lat, can_tas_histmn["tas"], transform=map, 
                     cmap="coolwarm", vmin=-40, vmax=40)

# Set spatial extent of plot
ax.set_extent(alaska_bbox, crs=ccrs.PlateCarree())

# Add coastline to define region
ax.add_feature(cfeature.COASTLINE)

# Add colorbar and label it
cbar = plt.colorbar(plot, ax=ax)
cbar.set_label("Temperature (C)")

# Add grid lines
gl = ax.gridlines(draw_labels=True, linestyle="--") 
gl.top_labels = False

# Add title
ax.set_title("Average Surface Air Temperature Mean Alaska (1990-2020) \n (CanESM5)", fontsize= 20)
plt.show()
```

### CanESM5 Future Projections map - SSP370

```python
# Define object containing PlateCarree projection
map = ccrs.PlateCarree()

# Make figure and axes
fig, ax = plt.subplots(figsize=(20, 10), subplot_kw={"projection": map})

# Plot
plot = ax.pcolormesh(can_tas_projmn.lon, can_tas_projmn.lat, can_tas_projmn["tas"], transform=map, 
                     cmap="coolwarm", vmin=-40, vmax=40)

# Set spatial extent of plot
ax.set_extent(alaska_bbox, crs=ccrs.PlateCarree())

# Add colorbar and label it
cbar = plt.colorbar(plot, ax=ax)
cbar.set_label("Temperature (C)")

# Add coastline to define region
ax.add_feature(cfeature.COASTLINE)

# Add grid lines
gl = ax.gridlines(draw_labels=True, linestyle="--") 
gl.top_labels = False

# Add title, show plot
ax.set_title("Projected Average Surface Air Temperature Alaska (2050-2100) \n (CanESM5 - SSP 370)", fontsize= 20)
plt.show()
```

### CESM historical map - SSP 370

```python
# Define object containing PlateCarree projection
map = ccrs.PlateCarree()

# Create figure and axes
fig, ax = plt.subplots(figsize=(20, 10), subplot_kw={"projection": map})

# Plot 
plot = ax.pcolormesh(cesm_tas_histmn.lon, cesm_tas_histmn.lat, cesm_tas_histmn["tas"], transform=map, 
                     cmap="coolwarm", vmin=-40, vmax=40)

# Set spatial extent of plot
ax.set_extent(alaska_bbox, crs=ccrs.PlateCarree())


# Add colorbar and label it
cbar = plt.colorbar(plot, ax=ax)
cbar.set_label("Temperature (C)")

# Add coastline to define region
ax.add_feature(cfeature.COASTLINE)

# Add grid lines
gl = ax.gridlines(draw_labels=True, linestyle="--") 
gl.top_labels = False

# Add title, show plot
ax.set_title("Average Surface Air Temperature Alaska (1990-2020) \n (CESM))", fontsize= 20)
plt.show()
```

```python
# Define object containing PlateCarree projection
map = ccrs.PlateCarree()

# Create figure and axes
fig, ax = plt.subplots(figsize=(20, 10), subplot_kw={"projection": map})

# Plot
plot = ax.pcolormesh(cesm_tas_projmn.lon, cesm_tas_projmn.lat, cesm_tas_projmn["tas"], transform=map, 
                     cmap="coolwarm", vmin=-40, vmax=40)

# Set spatial extent of plot
ax.set_extent(alaska_bbox, crs=ccrs.PlateCarree())


# Add colorbar and label it
cbar = plt.colorbar(plot, ax=ax)
cbar.set_label("Temperature (C)")

# Add coastline to define region
ax.add_feature(cfeature.COASTLINE)

# Add grid lines
gl = ax.gridlines(draw_labels=True, linestyle="--") 
gl.top_labels = False

# Add title, show plot
ax.set_title("Projected Average Surface Air Temperature Alaska (2050-2100) \n (CESM - SSP 370)", fontsize= 20)
plt.show()
```
